```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics

def plot_confusion_matrix(confusion_matrix, labels):
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=labels)
    cm_display.plot()
    plt.show()

    # Calculate accuracy
    accuracy = np.sum(np.diag(confusion_matrix)) / np.sum(confusion_matrix)

    # Calculate precision
    precision = np.diag(confusion_matrix) / np.sum(confusion_matrix, axis=0)

    # Calculate recall
    recall = np.diag(confusion_matrix) / np.sum(confusion_matrix, axis=1)

    return accuracy, precision, recall

def get_user_input():
    num_classes = int(input("Enter the number of classes (labels): "))
    labels = list(input("Enter the labels separated by space (e.g., 'A B C'): ").split())

    # Initialize an empty matrix
    confusion_matrix = np.zeros((num_classes, num_classes), dtype=int)

    print("You will now enter the confusion matrix values for each cell.")
    for i in range(num_classes):
        for j in range(num_classes):
            confusion_matrix[i, j] = int(input(f"Enter the count for Actual {labels[i]} Predicted {labels[j]}: "))

    return confusion_matrix, labels

def main():
    confusion_matrix, labels = get_user_input()
    accuracy, precision, recall = plot_confusion_matrix(confusion_matrix, labels)

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics

def plot_confusion_matrix(confusion_matrix, labels):
    # Choose a bright colormap, 'viridis' is a good choice for visibility
    cmap = plt.cm.viridis
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=labels)
    cm_display.plot(cmap=cmap)
    plt.show()

    # Calculate accuracy
    accuracy = np.sum(np.diag(confusion_matrix)) / np.sum(confusion_matrix)

    # Calculate precision
    precision = np.diag(confusion_matrix) / np.sum(confusion_matrix, axis=0)

    # Calculate recall
    recall = np.diag(confusion_matrix) / np.sum(confusion_matrix, axis=1)

    return accuracy, precision, recall

def get_user_input():
    num_classes = int(input("Enter the number of classes (labels): "))
    labels = list(input("Enter the labels separated by space (e.g., 'A B C'): ").split())

    # Initialize an empty matrix
    confusion_matrix = np.zeros((num_classes, num_classes), dtype=int)

    print("You will now enter the confusion matrix values for each cell.")
    for i in range(num_classes):
        for j in range(num_classes):
            confusion_matrix[i, j] = int(input(f"Enter the count for Actual {labels[i]} Predicted {labels[j]}: "))

    return confusion_matrix, labels

def main():
    confusion_matrix, labels = get_user_input()
    accuracy, precision, recall = plot_confusion_matrix(confusion_matrix, labels)

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)

if __name__ == "__main__":
    main()
```
