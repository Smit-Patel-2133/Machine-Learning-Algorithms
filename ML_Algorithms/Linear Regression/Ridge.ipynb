```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
n_samples, n_features = 100, 10
X = np.random.randn(n_samples, n_features)
w_true = np.random.randn(n_features)
y = X.dot(w_true) + 0.5 * np.random.randn(n_samples)
```

```python
train_size = int(n_samples * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
alpha = 0.1
ridge = Ridge(alpha=alpha)
ridge.fit(X_train, y_train)
```

```python
y_pred = ridge.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean squared error: {mse:.2f}")
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(y_test, y_pred, color='blue', label='True vs Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)  # Plotting the diagonal line
plt.scatter(y_test, y_test, color='red', label='True vs Original')  # Adding original values
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('True vs Predicted Values (Ridge Regression)')
plt.legend()
plt.grid(True)
plt.show()
```
